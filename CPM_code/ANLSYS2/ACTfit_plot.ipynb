```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('darkgrid')
```

```python
# load some CSVs 
# ACT
act_mean = pd.read_csv('150VT200P1800LowEP2_single_ACT.csv')
act_std = pd.read_csv('150VT200P1800LowEP2_single_ACT_std.csv')

# PRFDR
prfdr_mean = pd.read_csv('V150_PRFDR_single2.csv')
prfdr_std = pd.read_csv('V150_PRFDR_single2_std.csv')
```

```python
print(act_mean.columns)
print(prfdr_mean.columns)
```

```python
## add std to main : 
act_mean['std_speed'] = act_std.speed
act_mean['std_persistance'] = act_std.persistance

prfdr_mean['std_speed'] = prfdr_std.speed
prfdr_mean['std_persistance'] = prfdr_std.persistance
#act_mean.std_speed
```

```python
# ACT 
# plot lambda vs speed : 
plt.figure(figsize = (10,5))
sc = sns.scatterplot(x = "Lambda",y = "speed",size = "std_speed" ,data = act_mean)
axes = sc.axes
axes.set_ylim(0,0.0035)

# plot lambda vs Persist : 
plt.figure(figsize = (10,5))
sc = sns.scatterplot(x = "Lambda",y = "persistance",size = "std_persistance" ,data = act_mean)
#axes = sc.axes
#axes.set_ylim(0,0.0035)

# plot act vs speed : 
plt.figure(figsize = (10,5))
sc = sns.scatterplot(x = "Max_act",y = "speed",size = "std_speed" ,data = act_mean)
axes = sc.axes
axes.set_ylim(0,0.0035)

# plot act vs Persist : 
plt.figure(figsize = (10,5))
sc = sns.scatterplot(x = "Max_act",y = "persistance",size = "std_persistance" ,data = act_mean)
#axes = sc.axes
#axes.set_ylim(0,0.0035)
```

```python
# ACT 
# plot lambda vs speed : 
plt.figure(figsize = (10,5))
sc = sns.scatterplot(x = "Lambda",y = "speed",size = "std_speed" ,data = prfdr_mean)
#axes = sc.axes
#axes.set_ylim(0,0.0035)

# plot lambda vs Persist : 
plt.figure(figsize = (10,5))
sc = sns.scatterplot(x = "Lambda",y = "persistance",size = "std_persistance" ,data = prfdr_mean)
#axes = sc.axes
#axes.set_ylim(0,0.0035)

# plot act vs speed : 
plt.figure(figsize = (10,5))
sc = sns.scatterplot(x = "Persist",y = "speed",size = "std_speed" ,data = prfdr_mean)
axes = sc.axes
axes.set_xscale('symlog')
#axes.set_ylim(10**-5,10**-1)

# plot act vs Persist : 
plt.figure(figsize = (10,5))
sc = sns.scatterplot(x = "Persist",y = "persistance",size = "std_persistance" ,data = prfdr_mean)
axes = sc.axes
axes.set_xscale('log')
```

```python
fig, axs = plt.subplots(2,2)
# speed : 
axs[0,0].scatter(act_mean.Lambda,act_mean.speed)
axs[0,0].scatter(prfdr_mean.Lambda,prfdr_mean.speed)
axs[0,1].scatter(act_mean.Max_act,act_mean.speed)
axs[0,1].scatter(prfdr_mean.Persist,prfdr_mean.speed)
```
