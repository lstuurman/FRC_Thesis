```python
import plotly as py
import plotly.graph_objs as go
import networkx as nx
import numpy as np
from mayavi import mlab
import plotly.express as px
```

```python
g = nx.random_geometric_graph(1000,.15, dim=3, p=2, seed=None)
# extract postition directly from graph : 
pos = []
for n in g.nodes(data = True):
    #print(n)
    pos.append(n[1]['pos'])
    
xyz = np.array(pos)
x,y,z = xyz.T
degrees = [d[-1]/100 for d in g.degree()]
print(nx.is_connected(g))
```

```python
# Gc = max(nx.connected_components(g), key=len)
largest_cc = max(nx.connected_components(g), key=len)
Gc = g.subgraph(list(largest_cc))
#S = [g.subgraph(c).copy() for c in largest_cc]
print(nx.is_connected(Gc))
print(g.nodes()[0]['pos'])
print(Gc.nodes()[0]['pos'])
print(Gc.number_of_nodes())
```

```python
x_edges = []
y_edges = []
z_edges = []

for edge in g.edges():
#     print(edge)
#     print(g.nodes[edge[0]]['pos'])
    x0, y0, z0 = g.nodes[edge[0]]['pos']
    x1, y1, z1 = g.nodes[edge[1]]['pos']
#     x2, y2, z2 = g.nodes[edge[2]]['pos']
    x_edges.append(x0)
    x_edges.append(x1)
#     x_edges.append(x2)
    x_edges.append(None)
    y_edges.append(y0)
    y_edges.append(y1)
#     y_edges.append(y2)
    y_edges.append(None)
    z_edges.append(z0)
    z_edges.append(z1)
#     z_edges.append(z2)
    z_edges.append(None)
```

```python
trace1=go.Scatter3d(x=x_edges,
               y=y_edges,
               z=z_edges,
               mode='lines',
               line=dict(color='rgb(125,125,125)', width=1.5)
               )

# trace1 = go.Scatter3d(
#     x=x, y=y, z=z,
#     marker=dict(
#         size=6,
#         color=z,
#         colorscale='Viridis',
#     ),
#     line=dict(
#         color='#1f77b4',
#         width=1
#     )
# )

trace2=go.Scatter3d(x=x,
               y=y,
               z=z,
               mode='markers',
               marker=dict(symbol='circle',
                             size=7,
                             color=degrees,
                             colorscale='Viridis',
                             line=dict(color='rgb(50,50,50)', width=0.5)
                             )
               )

# trace2 = go.line_3d(x,y,z)

axis=dict(showbackground=False,
          showline=False,
          zeroline=False,
          showgrid=False,
          showticklabels=False,
          title=''
          )

layout = go.Layout(
         title="3D Representation of randomg geometric graph",
         width=1000,
         height=1000,
         showlegend=False,
         scene=dict(
             xaxis=dict(axis),
             yaxis=dict(axis),
             zaxis=dict(axis),
        ))#,
#      margin=dict(
#         t=100
#     ),
#     hovermode='closest',
#     annotations=[
#            dict(
#            showarrow=False,
#             xref='paper',
#             yref='paper',
#             x=0,
#             y=0.1,
#             xanchor='left',
#             yanchor='bottom',
#             font=dict(
#             size=14
#             )
#             )
#         ],    )
```

```python
data=[trace1,trace2]
fig=go.Figure(data=data, layout=layout)
fig.show()
```

```python
from mayavi import mlab
degrees = [d[-1]/100 for d in g.degree()]
mlab.close(all = True)
mlab.figure(bgcolor=(.8, .8, .8),size=(1000, 1000))
#plot lines first
# mlab.plot3d(x,y,z,color = (.5,.5,.5),
#             tube_radius=0.0037,
#             tube_sides = 32,
#             representation = 'wireframe')

pts = mlab.points3d(x,y,z,degrees,
             scale_factor = .2,
             scale_mode = 'scalar',
             resolution = 20)


pts.mlab_source.dataset.lines = np.array(list(g.edges()))
tube = mlab.pipeline.tube(pts, tube_radius=0.003)
mlab.pipeline.surface(tube, color=(0.5, 0.5, 0.5),opacity = .3)
```
