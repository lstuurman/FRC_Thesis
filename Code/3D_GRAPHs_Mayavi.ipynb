```python
import numpy as np
import networkx as nx
from mayavi import mlab
%gui qt
```

```python

# Random geometric graph:
g = nx.random_geometric_graph(200,.3, dim=3, p=2, seed=None)
# extract postition directly from graph : 
pos = []
for n in g.nodes(data = True):
    #print(n)
    pos.append(n[1]['pos'])
# print(pos)
# 3d positions : 
# pos = nx.spring_layout(g,dim = 3)
# pos = nx.random_layout(g,dim = 3)
# pos = nx.kamada_kawai_layout(g,dim = 3)

# pos = nx.shell_layout(g,dim=3)
# pos = nx.circular_layout(g,dim=3)
# pos = nx.spring_layout(g,dim=3)
# pos = nx.random_layout(g,dim=3)
# pos = nx.spectral_layout(g,dim=3)

# prepare x,y,z coordinates
#xyz = np.array([pos[key] for key in sorted(pos.keys())])
xyz = np.array(pos)
x,y,z = xyz.T
# scale plotted node sizes with degree
degrees = [d[-1]/100 for d in g.degree()]
mlab.close(all = True)
mlab.figure(bgcolor=(.8, .8, .8),size=(1000, 1000))
#plot lines first
# mlab.plot3d(x,y,z,color = (.5,.5,.5),
#             tube_radius=0.0037,
#             tube_sides = 32,
#             representation = 'wireframe')

pts = mlab.points3d(x,y,z,degrees,
             scale_factor = .2,
             scale_mode = 'scalar',
             resolution = 20)


pts.mlab_source.dataset.lines = np.array(list(g.edges()))
tube = mlab.pipeline.tube(pts, tube_radius=0.003)
mlab.pipeline.surface(tube, color=(0.5, 0.5, 0.5),opacity = .3)
```

```python
# Create the data.
from numpy import pi, sin, cos, mgrid
dphi, dtheta = pi/250.0, pi/250.0
[phi,theta] = mgrid[0:pi+dphi*1.5:dphi,0:2*pi+dtheta*1.5:dtheta]
m0 = 4; m1 = 3; m2 = 2; m3 = 3; m4 = 6; m5 = 2; m6 = 6; m7 = 4;
r = sin(m0*phi)**m1 + cos(m2*phi)**m3 + sin(m4*theta)**m5 + cos(m6*theta)**m7
x = r*sin(phi)*cos(theta)
y = r*cos(phi)
z = r*sin(phi)*sin(theta)

# View it.
from mayavi import mlab
s = mlab.mesh(x, y, z)
mlab.show()
```

```python
# needs mayavi2
# run with ipython -wthread
import networkx as nx
import numpy as np
from mayavi import mlab
mlab.options.offscreen = False

# some graphs to try
# H=nx.krackhardt_kite_graph()
# H=nx.Graph();H.add_edge('a','b');H.add_edge('a','c');H.add_edge('a','d')
# H=nx.grid_2d_graph(4,5)
H = nx.cycle_graph(20)

# reorder nodes from 0,len(G)-1
G = nx.convert_node_labels_to_integers(H)
# 3d spring layout
pos = nx.spring_layout(G, dim=3)
#print(pos)
# numpy array of x,y,z positions in sorted node order
xyz = np.array([pos[v] for v in sorted(G)])
#print(xyz)
# scalar colors
scalars = np.array(list(G.nodes())) + 5
print(scalars)
mlab.figure(1, bgcolor=(0, 0, 0))
mlab.clf()

pts = mlab.points3d(xyz[:, 0], xyz[:, 1], xyz[:, 2],
                    scalars,
                    scale_factor=0.1,
                    scale_mode='none',
                    colormap='Blues',
                    resolution=20)

pts.mlab_source.dataset.lines = np.array(list(G.edges()))
tube = mlab.pipeline.tube(pts, tube_radius=0.01)
mlab.pipeline.surface(tube, color=(0.8, 0.8, 0.8))

#mlab.savefig('mayavi2_spring.png')
#mlab.show() # interactive window
```
