```python
import pandas as pd
import numpy as np
import re
import seaborn as sns
import matplotlib.pyplot as plt
import glob
import pickle
sns.set_style('darkgrid')
sns.set(font_scale = 1.4)
```

```python
files = glob.glob("/home/lau/GIT/FRC_Thesis/results/Graphs_stuff/edge_lengths*.txt")
```

```python
rows = []
for f in files:
    lengths = np.loadtxt(f)
    tp = f.split('_')[-1][-6:-4]
    if tp not in ['WS','GM']:
        print(tp)
        lengths = lengths/4
    print(tp,np.average(lengths))
    for l in lengths:
        rows.append([tp,l])
```

```python
df = pd.DataFrame(rows,columns = ['type','length'])
f, (ax1, ax2, ax3,ax4,ax5) = plt.subplots(1,5,figsize = (15,4))
axs = (ax1, ax2, ax3,ax4,ax5)
for i,tp in enumerate(df.type.unique()):
#     plt.figure()             # <==================== here!
    sns.distplot(df[df.type == tp].length,hist = True, ax = axs[i])
    axs[i].set_title(tp)
plt.tight_layout()
```
