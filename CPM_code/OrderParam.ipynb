```python
import pandas as pd
import numpy as np
import re
import glob
import seaborn as sns
import matplotlib.pyplot as plt
import pickle
from itertools import product,permutations
from numpy.linalg import norm
from scipy.spatial.distance import euclidean
sns.set_style('darkgrid')
sns.set(font_scale = 1.4)
```

```python
 # load tracks : 
track_files = glob.glob('../data/real_cells/Mlab_track*.txt')[:3]
tracks = []
for f in track_files:
    track = np.loadtxt(f)
    tracks.append(track)

# convert to lists of vectors : 
def to_vecs(cell_track):
    vecs = []
    for i in range(80):
        point1 = cell_track[i]
        point2 = cell_track[i + 1]
        v1 = np.array(point2 - point1)
        vecs.append(v1)
    return np.array(vecs)

vec_tracks = np.array([to_vecs(t) for t in tracks])
for i in vec_tracks:
    print(i.shape)
```

```python
# calculate angles between same timestep of every cell:

angels_per_t = []

for i in range(len(vec_tracks[0])):
    vecs_at_t = vec_tracks[:,i]
    pairs = permutations(vecs_at_t)
    print(list(pairs))
    cosines = [np.dot(v1,v2)/(norm(v1) * norm(v2)) for (v1,v2) in pairs] #if list(v1) != list(v2)
    angels_per_t += cosines
```

```python
print(np.average(angels_per_t))
```
