```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
```

```python
# NON moving increasing pressure
df = pd.read_csv('rows.txt',sep = '\t',header = None,names = ["seeded_cells","free_voxels","T_volume"] )
df.seeded_cells = df.seeded_cells * 5
#x = np.loadtxt('size_matrix.txt')
with open('size_matrix.txt') as f:
    lines = f.read().splitlines()
    
averages = [np.mean([int(i) for i in x.split('\t')]) for x in lines]
```

```python
df['average_size'] = averages
```

```python
df.head()
```

```python
f, (ax1, ax2) = plt.subplots(1,2,figsize = (15,8))
sns.lineplot(x = 'seeded_cells',y = 'T_volume', data = df[df.seeded_cells < 3000], ax = ax1)
sns.lineplot(x = 'seeded_cells',y = 'average_size', data = df[df.seeded_cells < 3000], ax = ax2)
```

```python
# NON moving increasing pressure
df = pd.read_csv('rows_moving.txt',sep = '\t',header = None,names = ["seeded_cells","free_voxels","T_volume"] )
df.seeded_cells = df.seeded_cells * 5
#x = np.loadtxt('size_matrix.txt')
with open('matrix_moving.txt') as f:
    lines = f.read().splitlines()
print(df.count())
averages = [np.mean([int(i) for i in x.split('\t')]) for x in lines]
print(len(averages))
df['average_size'] = averages[:438]
```

```python
f, (ax1, ax2) = plt.subplots(1,2,figsize = (15,8))
sns.lineplot(x = 'seeded_cells',y = 'T_volume', data = df[df.seeded_cells < 3000], ax = ax1)
sns.lineplot(x = 'seeded_cells',y = 'average_size', data = df[df.seeded_cells < 3000], ax = ax2)
```
