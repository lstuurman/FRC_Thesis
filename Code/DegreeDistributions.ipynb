```python
d = {}

for gtype in df.type.unique():
    d[gtype] = []
    for i in range(10):
        bins = np.linspace(0,40,41)
        df1 = df[df.iter == i]
        df1 = df1[df1.type == gtype]
        print(gtype)
        print(i)
        print(len(df1.Degree.values))
        freqs,_,_ = plt.hist(df1.Degree,bins)
        plt.show()
        d[gtype].append(freqs)
        
# df1 = df[df.iter == 1]
# df1 = df1[df.type == 'power']
```
